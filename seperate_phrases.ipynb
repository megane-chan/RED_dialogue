```python
import pandas as pd
import re
```

```python
chat_data = pd.read_csv("data/chat.csv")
trans_data = pd.read_csv("data/trans.csv")

display(chat_data)
trans_data
```

```python
# Function to split text by punctuation and create new rows
def split_text_by_punctuation(row):
    sentences = re.split(r'(?<=[.!?,;])\s', row['content'])
    new_rows = []

    if len(sentences) > 1:
        for sentence in sentences:
            new_row = row.copy()
            new_row['content'] = sentence
            new_rows.append(new_row)
        return new_rows
    else:
        return [row]
```

```python
#split all the data by punctuation, so that each new utterance has its own row
trans_rows = []
for _, row in trans_data.iterrows():
    splits = split_text_by_punctuation(row)
    trans_rows = trans_rows + splits


chat_rows = []
for _, row in chat_data.iterrows():
    splits = split_text_by_punctuation(row)
    chat_rows = chat_rows + splits
```

```python
new_trans_data = pd.DataFrame(trans_rows)
new_chat_data = pd.DataFrame(chat_rows)

display(new_chat_data)
new_trans_data
```

```python
new_trans_data.to_csv("data/split_trans.csv")
new_chat_data.to_csv("data/split_chat.csv")
```
